```python
## bar chart to see the distribution of the column 'Age'

# your code goes here
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(df_survey['Age'], hist=True, kde=True,


              bins=int(180/5), color = 'darkblue',

              hist_kws={'edgecolor':'black'},
              kde_kws={'linewidth': 4})

plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency') 

plt.show()
```

```python
## bar chart to see the job posting distribution

# your code goes here
df_survey['Job'].value_counts().plot(kind='barh', figsize=(12, 12))

plt.xlabel('Count')
plt.ylabel('Job')
plt.title('Bar chart of Job')

plt.show()
```

```python
# show average anual salary data type with pandas
```

```python

df_survey['Average Annual Salary'].dtype
```

```python
# show the popular languages in bar chart using Language and Average Annual Salary in vertical bar chart using pink bar color

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# your code goes here
df_survey.groupby(['Language'])['Average Annual Salary'].mean().plot(kind='barh', color='pink')

plt.xlabel('Average Annual Salary')
plt.ylabel('Language')
plt.title('Bar chart of Average Annual Salary vs Language')

plt.show()

```
